## Important Rows : 
- genres
- keyword
- overview
- movie_id
- title
- cast
- crew

```python
movies_df = movies_df[['movie_id', 'genres', 'keywords', 'overview', 'title', 'cast', 'crew']]
```

```python
movies_df.head()
```

```python
def data_preprocess(string):
    l = []
    for i in eval(string):
        l.append(i['name'])
    return l
```

```python
movies_df['genres'] = movies_df['genres'].apply(data_preprocess)
```

```python
movies_df['keywords'] = movies_df['keywords'].apply(data_preprocess)
```

```python
def movie_cast_preprocessing(string):
    l = []
    counter = 0
    for i in eval(string):
        l.append(i['name'])
        counter += 1
        if counter == 3:
            break
    return l
```

```python
movies_df['cast'] = movies_df['cast'].apply(movie_cast_preprocessing)
```

```python
def fetch_director_name(string):
    l = []
    for i in eval(string):
        if i['job'] == 'Director':
            l.append(i['name'])
    return l
```

```python
movies_df['crew'] = movies_df['crew'].apply(fetch_director_name)
```

```python
print(type(movies_df['overview'][0]))
movies_df['overview'][0]
```

```python
movies_df.isna().sum()
```

```python
## As there are some missing values in the overview column, we are going to delete those rows
movies_df.dropna(subset=['overview'], inplace=True)
```

```python
movies_df.isna().sum()
```

```python
## Spliting the sentence in overview column
movies_df['overview'] = movies_df['overview'].apply(lambda x: x.split())
```

```python
movies_df.head()
```

```python
movies_df['genres'] = movies_df['genres'].apply(lambda x: [i.replace(' ', '') for i in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x: [i.replace(' ', '') for i in x])
movies_df['cast'] = movies_df['cast'].apply(lambda x: [i.replace(' ', '') for i in x])
movies_df['crew'] = movies_df['crew'].apply(lambda x: [i.replace(' ', '') for i in x])
```

```python
movies_df.head()
```

```python
movies_df['tags'] = movies_df['overview'] + movies_df['genres'] + movies_df['keywords'] + movies_df['cast'] + movies_df['crew']
```

```python
# Keeping, movie_id, title, tags. else removing
movies_df = movies_df[['movie_id', 'title', 'tags']]
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x: ' '.join(x))
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x: x.lower()) 
```

```python
movies_df['tags'][0]
```

```python
## taking words to their root form, example: activities - activity, as these well create useless vectors
import nltk
from nltk.stem.porter import PorterStemmer
ps  = PorterStemmer()

def convert_rootForm(string):
    l = []
    for word in string.split():
        l.append(ps.stem(word))
    
    return ' '.join(l)
```

```python
movies_df['tags'] = movies_df['tags'].apply(convert_rootForm)
```

```python
movies_df['tags'][0]
```

## Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(movies_df['tags']).toarray()
```

```python
cv.get_feature_names_out()[:1000]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
movies_df.head()
```

```python
recommend_films = sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
print(recommend_films)
```

```python
for movie_index in recommend_films:
    print(movies_df['movie_id'][movie_index[0]])    
```

```python
def recommend_movies(movie):
    movie_index = movies_df[movies_df['title'] == movie].index[0]
    movie_similarity = similarity[movie_index]
    movies = sorted(list(enumerate(movie_similarity)), reverse=True, key=lambda x:x[1])[1:6]
    recommended_movies = []
    for movie in movies:
        recommended_movies.append(movies_df['title'][movie[0]])
    return recommended_movies
```
